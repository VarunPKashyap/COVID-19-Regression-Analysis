### Determining the maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated['Maximum Infection Rate/Day'] = max_infection_rates
```

```python
corona_dataset_aggregated.
```

```python
corona_data = pd.DataFrame(corona_dataset_aggregated['Maximum Infection Rate/Day'])
```

```python
corona_data.head(188)
```
